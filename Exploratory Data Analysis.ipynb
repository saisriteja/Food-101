```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
import random

root_dir = 'data/food-101/images/'
rows, cols = 17, 6
fig, ax = plt.subplots(rows, cols, frameon=False, figsize=(15, 25))
classes = sorted(os.listdir(root_dir))
for i in range(rows):
    for j in range(cols):
        try:
            food_dir = classes[i*cols + j]
        except:
            break
        img = random.choice(os.listdir(root_dir + food_dir + '/'))
        img = root_dir + food_dir+'/'+img
        img = plt.imread(img)
        ax[i][j].imshow(img)
        ax[i][j].text(0, -20, food_dir, size=10, rotation=0, ha="left", va="top", 
                      bbox=dict(boxstyle="round", ec="black", fc="white"))
    plt.setp(ax, xticks=[], yticks=[])
```
